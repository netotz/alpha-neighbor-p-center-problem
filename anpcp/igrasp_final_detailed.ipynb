```python
from grasp_final import read_results, get_filenames

names = get_filenames()
names
```

```python
do_save = True
```

```python
import os

import pandas as pd
from IPython.display import display

from utils import format_latex_table
from grasp_final import FINAL_PATH

for name in names:
    results = read_results(name)
    grouped = results.groupby("instance i p alpha".split(), sort=False).mean().reset_index(level=0, drop=True)

    print(name)
    display(grouped)
    
    concated = pd.concat(
        {
            "OF": grouped.iloc[:, :3],
            "imp": grouped.iloc[:, 3:6],
            "upc": grouped.iloc[:, 8:],
            "time": grouped.iloc[:, 6],
            "last": grouped.iloc[:, 7],
        },
        axis=1,
    )
    if do_save:
        format_latex_table(concated, os.path.join(FINAL_PATH, f"table_{name}.tex"));
```
