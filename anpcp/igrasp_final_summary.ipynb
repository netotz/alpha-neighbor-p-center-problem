```python
from grasp_final import read_results, get_filenames

names = get_filenames()
names
```

```python
import pandas as pd

final = pd.concat(
    read_results(name) for name in names
)
```

```python
df_groupby = final.groupby("instance i p alpha".split(), sort=False)
grouped = df_groupby.mean()
grouped
```

```python
grouped[
    # (grouped["diff"] == grouped["min_imp"])
    # &
    # (grouped["last_imp"] > 0)
    # &
    (grouped["time"] >= 1800)
]
```

```python
instance_groupby = final.iloc[:, [0, 6, 10, 11]].groupby("instance", sort=False)

gb_diff = instance_groupby["diff"]
min_diff = gb_diff.min()
max_diff = gb_diff.max()
avg_diff = gb_diff.mean()
count_diff = gb_diff.apply(lambda d: (d > 0).sum())

avg_time = instance_groupby["time"].mean()
count_li = instance_groupby["last_imp"].apply(lambda d: (d > 0).sum())
```

```python
summary = pd.concat(
    {
        "min_diff": min_diff,
        "max_diff": max_diff,
        "avg_diff": avg_diff,
        "count_diff": count_diff,
        "avg_time": avg_time,
        "count_li": count_li,
    },
    axis=1
)
summary
```
