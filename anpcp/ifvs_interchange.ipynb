# FVS vs Standard Interchange

The following are preliminary experiments to see the results of Fast Vertex Substitution (FVS) performance against a straightforward implementation of the interchange local search heuristic.

```python
from models.instance import Instance
from models.solver import Solver

instance = Instance.random(100, 50)
solver = Solver(instance, 10, 2, True)
```

```python
solver.solution
```

```python
from copy import deepcopy
```

```python
solver.alpha = 3
solver.allocate_all()
solver.update_obj_func()
solver.solution
```

```python
solver_i = deepcopy(solver)
%timeit -r 1 -n 1 solver_i.interchange(True)
```

```python
solver_i.solution
```

```python
solver_fvs = deepcopy(solver)
solver_fvs.alpha_range = set(range(1, 5))
solver_fvs.allocate_all()
solver_fvs.update_obj_func()

%timeit -r 1 -n 1 solver_fvs.fast_vertex_substitution(True)
```

```python
solver_fvs.solution
```
