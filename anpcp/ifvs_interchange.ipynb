# FVS vs NI

The following are preliminary experiments to compare the results of the Fast Vertex Substitution (FVS) performance against a straightforward implementation of the interchange local search heuristic, called Naive Interchange (NI).

```python
from models.instance import Instance
from models.solver import generate_solvers

import os

datapath = os.path.abspath("..\\data")

instances50 = [Instance.read_json(os.path.join(datapath, f"anpcp_50_50_{i}.json")) for i in range(5)]
solvers50 = generate_solvers(instances50, (0.1, 0.25, 0.5), (2, 3))

instances100 = [Instance.read_json(os.path.join(datapath, f"anpcp_100_100_{i}.json")) for i in range(3)]
solvers100 = generate_solvers(instances100, (0.1, 0.25, 0.5), (2, 3))

instances500 = [Instance.read_json(os.path.join(datapath, f"anpcp_500_500_{i}.json")) for i in range(1)]
solvers500 = generate_solvers(instances500, (0.1, 0.25, 0.5), (2, 3))
```

```python
from utils import compare_local_search

results50 = compare_local_search(solvers50, True, False)
results50
```

```python
results100 = compare_local_search(solvers100, True, False)
results100
```

```python
results500 = compare_local_search(solvers500, True, False)
results500
```

```python
import pandas as pd

results = pd.concat([results50, results100, results500])
results
```

**How faster is FVS compared to NI?**

```python
def get_timediff(dataframe: pd.DataFrame) -> float:
    timediff = dataframe["NI", "time"] / dataframe["FVS", "time"]
    return timediff.mean()
```

$n = 50$

```python
get_timediff(results50)
```

$n = 100$

```python
get_timediff(results100)
```

$n = 500$

```python
get_timediff(results500)
```

```python
get_timediff(results)
```
