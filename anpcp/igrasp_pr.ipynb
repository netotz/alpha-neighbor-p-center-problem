```python
from grasp_pr import read_results, get_filenames

names = get_filenames()
names
```

```python
import pandas as pd

path_relinking = pd.concat(
    read_results(name) for name in names
)
```

```python
df_groupby = path_relinking.groupby("instance i p alpha algorithm pool_limit".split(), sort=False)
grouped = df_groupby.mean()
grouped
```

```python
instance_groupby = path_relinking.iloc[:, [0, 4, 5, 8, 9, 10, 11, 12, 13]].groupby("instance algorithm pool_limit".split(), sort=False)
summary = instance_groupby.mean()
summary
```
