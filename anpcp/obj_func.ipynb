# Understanding the objective function of the $\alpha$-neighbor _p_-center problem

The objective function of the problem is defined as:

$$
\mathcal{F}^\alpha (P) = \max\limits_{v \in V \setminus P}
        \min\limits_{S \subseteq P, |S| = \alpha}
            \max\limits_{s \in S}{d(s, v)}
$$

Where $V$ is the set of initial vertices, $P$ is the solution set, and $d(i,j)$ is the function of distance or dissimilarity from node $i$ to $j$.

The parameters for this demonstration will be:
* $n = 8$
* $p = 3$
* $\alpha = 2$

```python
n = 8
p = 3
alpha = 2
```

Let's generate random points:

```python
import nb_modules.obj_func_module as ofm

vertices = ofm.generate_randoms(n, 100, 100)
vertices
```

```python
import matplotlib.pyplot as plt

ofm.config_plot_points(vertices, 'gray')
plt.show()
```

And select a random solution, which is also feasible:

```python
import random

solution = random.sample(vertices, p)
solution
```

```python
clients = set(vertices) - set(solution)
ofm.config_plot_points(clients, 'skyblue')
ofm.config_plot_points(solution, 'red')
plt.show()
```

Clients are plotted in blue and facilities in red.

At each iteration, the current client is marked as a star, and the current solution subset of size $\alpha$ is marked as crosses. For each client, the distance to its $\alpha$-th closest facility is the orange line, i.e. its assigned facility.

The green line is the maximum distance of any client to its assigned facility, i.e. the objective function:

```python
ofm.eval_obj_func(alpha, clients, solution)
```
